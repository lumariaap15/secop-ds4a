```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
#cargamos los datos de Ejecucion de contratos
df_ejecucion = pd.read_csv("SECOP_II_-_Ejecuci_n_Contratos.csv", sep=',',low_memory=False)
df_ejecucion.head()
```

```python
df_ejecucion.dtypes
```

```python
#cargamos los datos de Contratos electrónicos
#creamos una lista de campos tipo fecha
date_cols = ['Ultima Actualizacion','Fecha Inicio Liquidacion', 'Codigo Proveedor','Fecha de Firma', 'Fecha de Inicio del Contrato',
       'Fecha de Fin del Contrato', 'Fecha de Inicio de Ejecucion',
       'Fecha de Fin de Ejecucion']
df_contratos_electronicos = pd.read_csv("SECOP_II_-_Contratos_Electr_nicos.csv", sep=',',low_memory=False, parse_dates=date_cols)
df_contratos_electronicos.head()
```

```python
#cantidad de registros cargados
df_contratos_electronicos.shape
```

```python
#tipo de datos en el dataframe
df_contratos_electronicos.dtypes
```

```python
#columnas del dataset
df_contratos_electronicos.columns
```

```python
#cantidad de contratos electrónicos que estan en el dataset de contratos en ejecución.
df_contratos_electronicos[df_contratos_electronicos['ID Contrato'].isin(df_ejecucion['IdentificadorContrato'])].shape
```

```python
#Gráfica de cantidad de contratos por estado
df_contratos_electronicos.groupby('Estado Contrato')['ID Contrato'].count().plot(kind='bar')
```

```python
#Quitamos los campos que no son relevantes para nuestro modelo 
df_contratos_electronicos = df_contratos_electronicos.drop(['Estado BPIN', 'Código BPIN', 'Anno BPIN',  'URLProceso',  'Nombre Representante Legal','Nacionalidad Representante Legal','Tipo de Identificación Representante Legal', 'Identificación Representante Legal',  'Género Representante Legal',  'Objeto del Contrato',  'Fecha Fin Liquidacion'], axis=1)
```

```python
#Cargamos el dataset de adiciones
df_adiciones = pd.read_csv("SECOP_II_-_Adiciones.csv", sep=',',low_memory=False)
df_adiciones.head() 
```

```python
#Numero de contratos electrónicos con dias adicionados
conadicion = df_contratos_electronicos[df_contratos_electronicos['Dias Adicionados']!=0]
conadicion.shape
```

```python
# Numero de contratos con dias adicionados por Estado
conadicion.groupby('Estado Contrato')['ID Contrato'].count().sort_values()
```

```python
#Número de contratos electrónicos que tienen adiciones
df_contratos_electronicos[df_contratos_electronicos['ID Contrato'].isin(df_adiciones['ID_Contrato'])].shape
```

```python
df_contratos_electronicos = df_contratos_electronicos.rename({'ID Contrato':'ID_Contrato'}, axis=1)
```

```python
result = pd.merge(df_contratos_electronicos, df_adiciones, on='ID_Contrato')
result.shape
```

## Statistical Visualizations

The first visualization explored is a pairplot. To be able to have a good representation and practicall usage of the kernel, this plot was done with a sample of 100 elements and only included the main variables identified on the data cleaning process.

```python
df_graphs=df.sample(100)
```

```python
df_graphs.shape
```

```python
# Plot Sizes

g=sns.pairplot(df_graphs, vars=[ 'Valor del Contrato', 'Dias Adicionados', 'Delay'] , diag_kind="hist")
g.fig.set_size_inches(10,4)
```
